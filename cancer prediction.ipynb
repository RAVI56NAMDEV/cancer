```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer(as_frame=True).frame
data.head()
```

```python
X,Y=data.iloc[:,:-1],data.iloc[:,-1]
X.head()
```

```python
import matplotlib.pyplot as plt

X.hist(figsize=(10,10))
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
import pandas as pd

scaler=StandardScaler()
minmax=MinMaxScaler()
x_scld=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
x_scld
```

```python
x_scld.hist(figsize=(10,10))
plt.tight_layout()
plt.show()
```

```python
from sklearn.manifold import TSNE

scores=pd.DataFrame(TSNE().fit_transform(x_scld),columns=['a','b'])
scores
```

```python
plt.scatter(scores.a,scores.b,c=Y)
plt.show()
```

```python
from sklearn.decomposition import PCA

pca=PCA(30)
pca.fit(x_scld)
```

```python
import numpy as np
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
pca=PCA(12)
pcs=pca.fit_transform(x_scld)
pcs.shape
```

```python
Y.shape
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(pcs,Y,test_size=.20)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree

dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
plt.figure(figsize=(15,10))
plot_tree(dtc)
plt.show()
```

```python
import sklearn
dir(sklearn.metrics)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,accuracy_score,classification_report,confusion_matrix

pred=dtc.predict(x_test)
pred
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.20)
est=[('scaler',StandardScaler()),('PCA',PCA()),('dtc',DecisionTreeClassifier())]

pipe=Pipeline(est)
pipe.fit(x_train,y_train)
pred=pipe.predict(x_test)
accuracy_score(y_test,pred)
```
